```java
//Carré

import java.util.Scanner;

int nbEtoile;
String choix;

Scanner clavier = new Scanner(System.in);

// Saisie taille carrée
do { 
    System.out.println("Saisissez une variable entre 2 et 15");
    nbEtoile = clavier.nextInt();

    if (nbEtoile < 2)
        System.out.println("La valeur doit être supérieure ou égale à 2");
    else if (nbEtoile > 15)
        System.out.println("La valeur doit être inférieure ou égale à 15");
}
while (nbEtoile < 2 || nbEtoile > 15);

// Plein/creux
System.out.println("Voulez-vous un carré plein ou creux ? (p/c)");
choix = clavier.next();

// Affichage du carré
for (int j = 0; j < nbEtoile; j++) {

    for (int i = 0; i < nbEtoile; i++) {

        if (choix.equals("c")) {
            // carré creux
            if (j == 0 || j == nbEtoile - 1 || i == 0 || i == nbEtoile - 1) {
                System.out.print("* ");
            } else {
                System.out.print("  ");
            }
        } else {
            // carré plein
            System.out.print("* ");
        }
    }
    System.out.println();
}
```
